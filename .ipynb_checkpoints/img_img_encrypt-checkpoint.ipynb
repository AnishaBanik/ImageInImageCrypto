```python
from PIL import Image
import numpy as np
import copy

#get cover image and form RGB array
im_c = Image.open("./"+"images.png")
width_c, height_c = im_c.size
np_c = np.asarray(im_c)
np_cover = copy.deepcopy(np_c)

R_c=[]
G_c=[]
B_c=[]

for i in range(len(np_cover[0])):
    R_c.append(np_cover[0][i][0])
    G_c.append(np_cover[0][i][1])
    B_c.append(np_cover[0][i][2])
print("cover: ", len(np_cover[0]))
#print(B_c)

#get secret image and resize it, save it, and then form RGB arrays of it
im_s = Image.open("./"+"sign.png")
new_image = im_s.resize((int(width_c/8), int(height_c/8)))
new_image.save('small_secret.png')
im_ss = Image.open("./"+"small_secret.png")
width_s, height_s = im_ss.size
np_s = np.asarray(im_ss)
np_secret = copy.deepcopy(np_s)

R_s=[]
G_s=[]
B_s=[]

for i in range(len(np_secret[0])):
    R_s.append(np_secret[0][i][0])
    G_s.append(np_secret[0][i][1])
    B_s.append(np_secret[0][i][2])
print("cover: ", len(np_secret[0]))
#print(R_s)


#"before ",bin_cov_val
#"insert ", bin_sec_val[k]
#"after ",bin_cov_val
j=0
for i in range(len(np_secret[0])):
    sec_val=R_s[i]
    bin_sec_val= bin(sec_val)[2:].zfill(8)
    for k in range(8):
        cov_val=R_c[j]
        bin_cov_val= bin(cov_val)[2:].zfill(8)
        bin_cov_val=bin_cov_val[0:7]+bin_sec_val[k]
        int_cov_val=int(bin_cov_val, 2)
        R_c[j]=int_cov_val
        j=j+1
        
j=0
for i in range(len(np_secret[0])):
    sec_val=G_s[i]
    bin_sec_val= bin(sec_val)[2:].zfill(8)
    for k in range(8):
        cov_val=G_c[j]
        bin_cov_val= bin(cov_val)[2:].zfill(8)
        bin_cov_val=bin_cov_val[0:7]+bin_sec_val[k]
        int_cov_val=int(bin_cov_val, 2)
        G_c[j]=int_cov_val
        j=j+1
        
j=0
for i in range(len(np_secret[0])):
    sec_val=B_s[i]
    bin_sec_val= bin(sec_val)[2:].zfill(8)
    for k in range(8):
        cov_val=B_c[j]
        bin_cov_val= bin(cov_val)[2:].zfill(8)
        bin_cov_val=bin_cov_val[0:7]+bin_sec_val[k]
        int_cov_val=int(bin_cov_val, 2)
        B_c[j]=int_cov_val
        j=j+1
        
#creating final cover image
for i in range(len(np_secret[0])):
    np_cover[0][i][0]=R_c[i]
    np_cover[0][i][1]=G_c[i]
    np_cover[0][i][2]=B_c[i]

im = Image.fromarray(np_cover)
im.save('dummy99.png')
print("complete")

print("-----------")        
#print(B_c)        
print("done")
```
