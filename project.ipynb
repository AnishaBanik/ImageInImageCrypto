```python
from PIL import Image
import numpy as np

im_frame = Image.open('./' + 'this1small.png')
np_frame = np.array(im_frame.getdata())
# img = Image.fromarray(np_frame) 
# img.save('dummy.png')
#extracting RGB elements
R=[]
G=[]
B=[]

for i in range(len(np_frame)):
    R.append(np_frame[i][0])
    G.append(np_frame[i][1])
    B.append(np_frame[i][2])
    
#converting first 9 into binary
for i in range(9):
    R[i]=format(np_frame[i][0],'08b')
    G[i]=format(np_frame[i][1],'08b')
    B[i]=format(np_frame[i][2],'08b')


#creating LSB matrix 3*9
LSB = [[0 for x in range(9)] for y in range(3)] 
for i in range(9):
    LSB[0][i]=R[i][7]
for i in range(9):
    LSB[1][i]=G[i][7]
for i in range(9):
    LSB[2][i]=B[i][7]


#embedding secret inside LSB
secret="110111111101010000100001101"
k=0
for x in range(0,3):
    for y in range(0,9):
        LSB[x][y]=secret[k]
        k+=1

#putting embedded LSBs into RGB arrays on first 9 index
i=0
for x in range(0,9):
    R[x]=R[x][0:7]+LSB[0][i]
    G[x]=G[x][0:7]+LSB[1][i]
    B[x]=R[x][0:7]+LSB[2][i]
    i+=1
    
#converting arrays from binary to decimal
for x in range(0,9):
    R[x]=int(R[x],2)
    G[x]=int(G[x],2)
    B[x]=int(B[x],2)
    
#appending values from np_frame to make mini arrays to bigger ones
# for x in range(9,len(np_frame)):
#     R.append(np_frame[x][0])
#     G.append(np_frame[x][1])
#     B.append(np_frame[x][2])

#creating final np_frame
for i in range(9):
    np_frame[i][0]=R[i]
    np_frame[i][1]=G[i]
    np_frame[i][2]=B[i]

arr1 = np_frame.reshape(25, 25,-1)
print("---------")
print(arr1)
im = Image.fromarray(arr1)
im.save('dummy22.png')



# rgbArray = np.zeros((len(np_frame),len(np_frame),3), 'uint8')
# for i in range(len(np_frame)):
#     rgbArray[i][0]=R[i]*255
#     rgbArray[i][1]=G[i]*255
#     rgbArray[i][2]=B[i]*255
# # rgbArray[..., 0] = R
# # rgbArray[..., 1] = G
# # rgbArray[..., 2] = B
# img = Image.fromarray(rgbArray,mode="RGB") 
# img.save('myimg1.png')

print("complete")
```
